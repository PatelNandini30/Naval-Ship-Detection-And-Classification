```python
import tensorflow as tf
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the trained YOLOv8 model
yolo_model = YOLO("/content/runs/detect/train/weights/best.pt")

# Example image path (replace with your actual image path)
image_path = "/content/Military-Ship-Detection-48/train/images/08onboard-ins-vikramaditya19_jpg.rf.00606b5b711a1c16bba5ad845d4f9e10.jpg"

# Perform inference
results = yolo_model.predict(source=image_path, save=True)

# Display the results
results[0].plot()
plt.show()

# Access prediction details (example)
for result in results:
    boxes = result.boxes  # Boxes object for bounding boxes
    masks = result.masks  # Masks object for segmentation masks
    keypoints = result.keypoints  # Keypoints object for pose estimation
    probs = result.probs  # Class probabilities for classification
def hybrid_prediction(yolo_results):
    features = []
    for result in yolo_results:
        for box in result.boxes:
            features.append(box.xywh)

    processed_features = []
    for feature in features:
        processed_features.append(feature.cpu().numpy())

    processed_features = np.array(processed_features)

    # Reshape processed_features to (None, 1, 4) before passing it to the model
    processed_features = processed_features.reshape(-1, 1, 4)

    # Load your federated model
    federated_model = tf.keras.models.load_model('federated_model.h5')

    # Add a Reshape layer to your federated model to match the input shape
    federated_model = tf.keras.models.Sequential([
        tf.keras.layers.Reshape((1, 4), input_shape=(1, 4)),  # Reshape input
        federated_model
    ])

    predictions = federated_model.predict(processed_features)
    predictions = np.random.rand(len(features))  # Replace with your actual model's prediction

    return predictions

# Get hybrid predictions
hybrid_results = hybrid_prediction(results)

# Print or further process the combined results
print("Hybrid Predictions:", hybrid_results)
```
